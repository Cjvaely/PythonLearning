# requests库讲解

## 实例引入

```python
import requests

response = requests.get('http://www.baidu.com/')
print('type(response) = ', type(response))
print('response.status_code = ', response.status_code)
print('type(response.text) = ', type(response.text))
print('response.text = ', response.text)
print('response.cookies = ', response.cookies)
```

## 请求

```python
import requests

requests.post('http://httpbin.org/post')
requests.put('http://httpbin.org/put')
requests.delete('http://httpbin.org/delete')
requests.head('http://httpbin.org/get')
requests.options('http://httpbin.org/get')
```

### 基本GET请求

#### 基本写法

```python
import requests

response = requests.get('http://httpbin.org/get')
print(response.text)
```

#### 带参数GET请求

```python
import requests

response = requests.get('http://httpbin.org/get?name=germey&age=22')
print(response.text)
```

```python
import requests

# 参数
data = {
    'name': 'germey',
    'age': 22
}
response = requests.get('http://httpbin.org/get', params=data)
print(response.text)
```

#### 解析json

```python
import requests

response = requests.get('http://httpbin.org/get')
# <class 'str'>
print(type(response.text))

print(response.json())
# <class 'dict'>
print(type(response.json()))
```

#### 获取二进制数据

```python
import requests

response = requests.get('https://github.com/favicon.ico')
# <class 'str'> <class 'bytes'>
print(type(response.text), type(response.content))
print(response.text)
print(response.content)
```

```python
import requests

response = requests.get('https://www.github.com/favicon.ico')
# 下载图片
with open('./imgs/favicon.ico', 'wb') as f:
    f.write(response.content)
    f.close()
```

### 基本POST请求

```python
import requests

# 表单
data = {
    'name':'germey',
    'age':'22'
}
response = requests.post('http://httpbin.org/post', data=data)
print(response.text)
```

## 响应

### response属性

```python
import requests

headers = {
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'
}
response = requests.get('https://www.jianshu.com/', headers=headers)
print('type(response.status_code) = ', type(response.status_code), response.status_code)
print('type(response.headers) = ', type(response.headers), response.headers)
print('type(response.cookies) = ', type(response.cookies), response.cookies)
print('type(response.url) = ', type(response.url), response.url)
print('type(response.history) = ', type(response.history), response.history)
```

### 状态码判断

```python
import requests

headers = {
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'
}
response = requests.get('https://www.jianshu.com')
exit if not response.status_code == requests.codes.ok else print('Response Successfully')
```

## 高级操作

### 文档上传

```python
import requests

files = {
    'file':open('./imgs/favicon.ico', 'rb')
}
response = requests.post('http://httpbin.org/post', files=files)
print(response.text)
```

### 获取cookie

```python
import requests

response = requests.get('https://www.baidu.com')
print(response.cookies)
for key, value in response.cookies.items():
    print(key + '=' + value)
```

### 会话维持

```python
import requests

# 一个浏览器访问网站，设置cookie
requests.get('http://httpbin.org/cookies/set/number/123456789')
# 另一个浏览器访问cookies
response = requests.get('http://httpbin.org/cookies')
print(response.text)
```

```python
import requests

# 开启会话，实现同一浏览器的cookies的set和get
s = requests.Session()
s.get('http://httpbin.org/cookies/set/number/123456789')
response = s.get('http://httpbin.org/cookies')
print(response.text)
```

### 证书验证

```python
import requests

# 网站证书不合法
response = requests.get('https://222.197.164.231/')
print(response.status_code)
```

```python
import requests

# 网站证书不合法
response = requests.get('https://222.197.164.231/')
print(response.status_code)
```

```python
import requests
from requests.packages import urllib3

# 消除警告信息（提醒检测证书）
urllib3.disable_warnings()
# 不检测证书
response = requests.get('https://222.197.164.231', verify=False)
print(response.status_code)
```

```python
import requests

# 设置特定证书检测
response = requests.get('https://222.197.164.231/', cert=('/path/server.crt', '/path/key'))
print(response.status_code)
```
